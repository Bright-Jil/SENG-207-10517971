```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("red_wine.csv")
```

```python
plt.bar(df['pH'], df['quality'])
plt.xlabel('pH') 
plt.ylabel('Quality')
plt.title('pH Against Wine Quality')
```

```python
plt.bar(df['alcohol'], df['quality'])
plt.xlabel('alcohol') 
plt.ylabel('quality')
plt.title('Alcohol Against Wine Quality')
```

```python
plt.bar(df['residual sugar'], df['quality'])
plt.xlabel('Residual Sugar') 
plt.ylabel('Quality')
plt.title('Residual Sugar Against Wine Quality')
```

The data set contains the various constituents of red wine. In the above codes, we see the relationships between the various components of the red wine. The bar chart especially was used to present a graphic relationship between some of the components of the red wine and the quality. This was to investigate the role each of these components play in the overall quality of the red wine. 

With regards to the pH, we see from the chart that moderate pH of 3.2 to 3.5 leads to a high quality wine while too low or too high pH's leads to low quality wine. Also, the quality of the redwine is improved when the alcohol content is moderate.

However, the bar chart also shows that the less the residual sugar the better the red wine.
